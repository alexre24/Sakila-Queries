```python
q = """SELECT 
    customer.first_name,
    customer.last_name,
    customer.email,
    customer.address_id,
    city.city
FROM
    customer
        JOIN
    city ON customer.customer_id = city.city_id
WHERE
    city_id = 312;"""
pd.read_sql(q, engine)
```

```python
q = """SELECT 
    film.title,
    film.description,
    film.release_year,
    film.rating,
    film.special_features
FROM
    film
        JOIN
    category ON film.film_id = category.category_id
WHERE
    category.name = 'Comedy';"""
pd.read_sql(q, engine)
```

```python
q = """SELECT 
    actor.last_name, film.title, film.release_year
FROM
    film
        JOIN
    actor ON film.film_id = actor.actor_id
WHERE
    actor.last_name = 'Lollobrigida';"""
pd.read_sql(q, engine)
```

```python
q = """SELECT 
    actor.first_name, actor.last_name, film.title
FROM
    actor
        JOIN
    film ON actor.actor_id = film.film_id
WHERE
    film.title = 'Bingo Talented';"""
pd.read_sql(q, engine)
```

```python
q = """SELECT 
    customer.customer_id, payment.amount
FROM
    payment
        JOIN
    customer ON payment.customer_id = customer.customer_id
GROUP BY customer.customer_id , payment.amount
HAVING payment.amount > (2 * 4.200667);"""
pd.read_sql(q, engine)
```

```python
q = """SELECT 
    customer.first_name,
    customer.last_name,
    COUNT(payment.payment_id) AS num_payments
FROM
    payment
        JOIN
    customer ON customer.customer_id = payment.customer_id
GROUP BY customer.first_name , customer.last_name
ORDER BY num_payments DESC
LIMIT 5;"""
pd.read_sql(q, engine)
```
